## Proporcions fenotípiques d'un creuament trihíbrid

Assumim tres gens no lligats (amb segregació independent), cadascun amb un al·lel dominant i un al·lel recessiu, i que determinen tres caràcters independents:

- **A** (dominant) determina el color groc del fruit i **a** (recessiu), el verd.
- **B** (dominant) determina la pell llisa i **b** (recessiu), rugosa.
- **C** (dominant) determina el color violeta de la flor i **c** (recessiu), el blanc.

Com que hi ha tres caràcters amb dos fenotips possibles en cada un, en total poden haver $2^3 = 8$ combinacions fenotípiques possibles:

  | Genotips | Fenotips                                  |
  | -------- | ----------------------------------------- |
  | A_ B_ C_ | Fruits grocs, pell llisa, flors violeta   |
  | A_ B_ cc | Fruits grocs, pell llisa, flors blanques  |
  | A_ bb C_ | Fruits grocs, pell rugosa, flors violeta  |
  | aa B_ C_ | Fruits verds, pell llisa, flors violeta   |
  | A_ bb cc | Fruits grocs, pell rugosa, flors blanques |
  | aa B_ cc | Fruits verds, pell llisa, flors blanques  |
  | aa bb C_ | Fruits verds, pell rugosa, flors violeta  |
  | aa bb cc | Fruits verds, pell rugosa, flors blanques |

La qüestió és determinar la proporció de cada **fenotip** en un creuament trihíbrid: **Aa Bb Cc** $\times$ **Aa Bb Cc**.

### Expansió del binomi de Newton

Els tres caràcters s'hereten independentment, i a més en la descendència d'un creuament trihíbrid cada caràcter ha de presentar un patró de segregació amb les proporcions 3:1 per als fenotips dominant i recessiu. És a dir, el fenotip total és la *combinació* de tres experiments de Bernoulli independents, amb probabilitats $3/4$ i $1/4$ per als dos resultats possibles (fenotip dominant o recessiu). Una manera d'obtenir expressions matemàtiques de les probabilitats dels 8 fenotips possibles és amb una expansió del binomi de Newton:

<p style="text-align: center;">
$(p + q)^3 = \binom{3}{0}p^3q^0 + \binom{3}{1}p^2q^1 + \binom{3}{2}p^1q^2 + \binom{3}{3}p^0q^3 = 1$
</p>

<p style="text-align: center;">
    $(p + q)^3 = p^3 + 3p^2q + 3pq^2 + q^3 = 1$
</p>

Si $p = 3/4$ és la probabilitat del fenotip dominant en la descendència del creuament trihíbrid, i $q=1/4$, la del fenotip recessiu, aleshores, els quatre termes de l'expansió anterior són:

<p style="text-align: center;">
    $\left( \frac{3}{4} \right)^3 + 3\cdot \left( \frac{3}{4} \right)^2 \frac{1}{4} + 3\cdot \frac{3}{4}\left( \frac{1}{4} \right)^2 + \left( \frac{1}{4} \right)^3$
</p>

Que corresponen a les quatre probabilitats binomials d'obtenir: els tres fenotips dominants, dos fenotips dominants i un recessiu, un fenotip recessiu i dos dominants, i finalment, els tres fenotips recessius:

  | Genotips | Fenotips                                  | Probabilitat  |
  | -------- | ----------------------------------------- | ------------- |
  | A_ B_ C_ | Fruits grocs, pell llisa, flors violeta   | 27/64         |
  | A_ B_ cc | Fruits grocs, pell llisa, flors blanques  |  9/64         |
  | A_ bb C_ | Fruits grocs, pell rugosa, flors violeta  |  9/64         |
  | aa B_ C_ | Fruits verds, pell llisa, flors violeta   |  9/64         |
  | A_ bb cc | Fruits grocs, pell rugosa, flors blanques |  3/64         |
  | aa B_ cc | Fruits verds, pell llisa, flors blanques  |  3/64         |
  | aa bb C_ | Fruits verds, pell rugosa, flors violeta  |  3/64         |
  | aa bb cc | Fruits verds, pell rugosa, flors blanques |  1/64         |
